```python
import cv2
import matplotlib.pyplot as plt
import numpy as np 
```

```python
# --- Step 1: Load the image ---
img = cv2.imread("images/dog.png")
```

```python
# --- Step 2: Convert to grayscale ---

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# --- Step 3: Apply histogram equalization ---
equalized = cv2.equalizeHist(gray)
```

```python
plt.figure(figsize=(12, 4))

# converting opencv to matplotlib
plt.subplot(1,3,1)
plt.axis("off")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original")

# grayscale plot
plt.axis("off")
plt.subplot(1,3,2)
plt.title("GRAYSCALE")
plt.imshow(gray, cmap="gray")

# eualized plot
plt.axis("off")
plt.subplot(1,3,3)
plt.title("Equalized")
plt.imshow(equalized, cmap="gray")
plt.tight_layout
plt.show()
```

```python
# --- Step 4: Plot the histogram ---
# this will plot the histogram of the original image, you can also plot the histogram of the equalized image by replacing img with equalized
plt.hist(img.ravel(), 256, [0,256]) # plot histogram
plt.title("Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()

plt.hist(equalized.ravel(), 256, [0,256]) # the ravel() function flattens the image array into a 1D array, which is required for the histogram function
#  also, the 256 is the number of bins for the histogram, and [0,256] is the range of pixel intensity values (0 to 255)

# bins are the intervals that the pixel intensity values are grouped into for the histogram. In this case, we have 256 bins, which means that each bin will represent a range of pixel intensity values (e.g., 0-1, 1-2, ..., 254-255). The histogram will count how many pixels fall into each bin, giving us a frequency distribution of pixel intensities in the image.
plt.title(" Equalized Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()
```
