## Sobel Operation
#### Sobel is a dicrete differentiation operation that computes the first deriviation of the image intensity function

```python
img2 = cv2.imread("images/image.jpg")
```

```python
# Apply Gaussian Blur to reduce noise
blurred = cv2.GaussianBlur(img2, (3,3), 0)

sobelx =cv2.Sobel(blurred, cv2.CV_64F, 1,0, ksize=3)
sobely = cv2.Sobel(blurred,cv2.CV_64F, 0,1, ksize=3)

```

```python
# Convert to absolute scale
abs_sobelx = cv2.convertScaleAbs(sobelx)
abs_sobely = cv2.convertScaleAbs(sobely)
```

```python
# Combine both directions
sobel_combine = cv2.addWeighted(abs_sobelx, 0.5, abs_sobley,0.5, 0)
```

```python
# Show results
cv2.imshow('Original Image', img)
cv2.imshow('Sobel X', abs_sobelx)
cv2.imshow('Sobel Y', abs_sobely)
cv2.imshow('Sobel Combined', sobel_combine)
cv2.waitKey(0)
```

```python
# Final display for Lab 4
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
sobel_magnitude = cv2.magnitude(sobelx, sobely)
sobel_magnitude = cv2.convertScaleAbs(sobel_magnitude)

plt.figure(figsize=(16, 4))

plt.subplot(1, 4, 1)
plt.imshow(img2_rgb)
plt.title("Original")
plt.axis("off")

plt.subplot(1, 4, 2)
plt.imshow(abs_sobelx, cmap="gray")
plt.title("Sobel X")
plt.axis("off")

plt.subplot(1, 4, 3)
plt.imshow(abs_sobely, cmap="gray")
plt.title("Sobel Y")
plt.axis("off")

plt.subplot(1, 4, 4)
plt.imshow(sobel_magnitude, cmap="gray")
plt.title("Sobel Magnitude")
plt.axis("off")

plt.tight_layout()
plt.show()
```
