### WEEK 2 Sampling And Qunatization

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load image
img = cv2.imread("images/image2.jpg")
```

```python
# Inspect the matrix structure
print("Image Shape", img.shape)
print ("Image Shape: ", img.shape)
print("Data type: ", img.dtype)
print("Min Value: ", img.min())
print("Max Value: ", img.max())
print("Mean Value: ", img.mean())
print("Standard Deviation: ", img.std())
print("Median Value: ", np.median(img))
print("Unique Values: ", np.unique(img))
```

```python
# Access a single pixel (row=100, col=200)
pixel = img[100, 200]
print("Pixel Value at (100, 200): ", pixel)

```

```python
# Convert to grayscale and inspect
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# Demonstrate different bit depths (quantization)
bit4 = (gray // 16) * 64
bit2 = (gray // 64) * 64

print("bit4 Size: ", bit4.size)
print("bit2 Size: ", bit2.size)
```

```python
# Plot comparison of bit depths
plt.figure(figsize=(12, 4))
plt.subplot(1,3,1); plt.axis("off"); plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)); plt.title("Original")

plt.subplot(1,3,2); plt.axis("off"); plt.imshow(gray, cmap="gray"); plt.title("Grayscale")

plt.subplot(1,3,3); plt.axis("off"); plt.imshow(bit4, cmap="gray"); plt.title("bit4")


plt.tight_layout()
plt.show()
```
