```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("images/image2.jpg")

if img is None:
    print("No image in such directory found")
```

```python
blurred = cv2.blur(img, (5,5))
cv2.imshow("Original", img)
cv2.imshow("Blurred", blurred)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
